```python
import pandas as pd
import numpy as np
pd.set_option('max_columns', 30)

trends = pd.read_csv('trends.csv')
trends.columns = ['time', 'trend']

ETH = pd.read_csv('ETH_price.csv')
ETH.drop(['open','high','low','volume','marketcap'],axis = 1, inplace = True)
ETH.columns = ['time', 'ETH_price']

train_data = pd.read_csv('train_dense_new.csv')
train_data['time'] = train_data['time'].apply(lambda x: x[:10])
```

```python
updated = pd.merge(train_data, trends, on = 'time')
updated = pd.merge(updated, ETH, on = 'time')
updated
```
